## Load required packages

```r
suppressMessages(require(Seurat))
suppressMessages(require(Matrix))
suppressMessages(require(data.table))
```

## Read in Reference and Braak 6 normalized counts

```r
system('tar -xf ../../results/SEA_AD/MTG/mtg_X_counts_ref_br6.tar.gz')
counts_ref_br6 <- fread('mtg_X_counts_ref_br6',header=TRUE)
system('rm mtg_X_counts_ref_br6')
```

```r
counts_ref_br6
```

## Transpose the data to make genes rows, as required by Seurat

```r
counts_ref_br6=transpose(counts_ref_br6)
```

```r
counts_ref_br6
```

## Create multiple seurat objects from the data, for memory issues

```r
seu1 <- CreateSeuratObject(counts=counts_ref_br6[,0:50000])
seu2 <- CreateSeuratObject(counts=counts_ref_br6[,50001:100000])
seu3 <- CreateSeuratObject(counts=counts_ref_br6[,100001:150000])
seu4 <- CreateSeuratObject(counts=counts_ref_br6[,150001:200000])
seu5 <- CreateSeuratObject(counts=counts_ref_br6[,200001:250000])
seu6 <- CreateSeuratObject(counts=counts_ref_br6[,250001:322769])
```

## Get rid of the data objects no longer needed

```r
# remove all objects that will not be used.
rm(counts_ref_br6)

# run garbage collect to free up memory
gc()
```

## Merge the various seurat objects

```r
alldata <- merge(seu1, c(seu2, seu3, seu4,seu5,seu6))
```

```r
# remove all objects that will not be used.
rm(seu1, seu2, seu3, seu4, seu5, seu6)

# run garbage collect to free up memory
gc()
```

## Normalize and Scale

```r
alldata <- NormalizeData(alldata)
all.genes <- rownames(alldata)
alldata <- ScaleData(alldata, features = all.genes)
```

## Read in metadata for cell identification as control or AD

```r
meta <- fread('../../results/SEA_AD/meta_ref_br6')
```

```r
alldata
```

## Add metadata to the merged seurat object

```r
alldata@meta.data <- meta
```

```r
alldata
```

```r
head(alldata)
```

```r
as.data.frame(alldata@assays$RNA@counts[0:10,])
```

## Set cell identifier as the metadata column

```r
Idents(alldata = alldata) <- alldata@meta.data$'condition'
```

## Find differentially expressed genes in AD and Control samples

```r
de_genes <- FindMarkers(alldata, ident.1 = "AD", ident.2 = "C")
```

```r
de_genes
```

```r
write.table(de_genes, "../../results/SEA_AD/X_de_genes_seurat.tsv", sep="\t")
```

## Repeat the procedure for UMI counts data

```r
counts_ref_br6 <- fread('../../results/SEA_AD/umi_ref_br6',header=TRUE)
counts_ref_br6 <- transpose(counts_ref_br6)
seu1 <- CreateSeuratObject(counts=counts_ref_br6[,0:50000])
seu2 <- CreateSeuratObject(counts=counts_ref_br6[,50001:100000])
seu3 <- CreateSeuratObject(counts=counts_ref_br6[,100001:150000])
seu4 <- CreateSeuratObject(counts=counts_ref_br6[,150001:200000])
seu5 <- CreateSeuratObject(counts=counts_ref_br6[,200001:250000])
seu6 <- CreateSeuratObject(counts=counts_ref_br6[,250001:322769])
rm(counts_ref_br6)
gc()
alldata <- merge(seu1, c(seu2, seu3, seu4,seu5,seu6))
rm(seu1, seu2, seu3, seu4, seu5, seu6)
gc()
alldata <- NormalizeData(alldata)
all.genes <- rownames(alldata)
alldata <- ScaleData(alldata, features = all.genes)
meta <- fread('../../results/SEA_AD/meta_ref_br6')
alldata@meta.data <- meta
Idents(alldata = alldata) <- alldata@meta.data$'condition'
de_genes <- FindMarkers(alldata, ident.1 = "AD", ident.2 = "C")
de_genes
```

```r
de_genes
```

```r
write.table(de_genes, "../../results/SEA_AD/counts_de_genes_seurat.tsv", sep="\t")
```
