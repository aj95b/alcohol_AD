```r
suppressMessages(require(Seurat))
suppressMessages(require(Matrix))
suppressMessages(require(data.table))
suppressMessages(require(DESeq2))
#library( "DESeq2" )
```

```r
counts_astrocytes <- fread('../../results/SEA_AD/MTG/counts_astrocytes_ref_br6.tsv', header = TRUE)
counts_microglia <- fread('../../results/SEA_AD/MTG/counts_microglia_ref_br6.tsv', header = TRUE)
counts_astrocytes <- transpose(counts_astrocytes)
counts_microglia <- transpose(counts_microglia)
meta_astrocytes <- fread('../../results/SEA_AD/MTG/meta_astrocytes_ref_br6.tsv', header = TRUE)
meta_microglia <- fread('../../results/SEA_AD/MTG/meta_microglia_ref_br6.tsv', header= TRUE)
seu_astrocytes <- CreateSeuratObject(counts=counts_astrocytes)
seu_microglia <- CreateSeuratObject(counts=counts_microglia)
seu_astrocytes <- NormalizeData(seu_astrocytes)
seu_astrocytes.genes <- rownames(seu_astrocytes)
seu_astrocytes <- ScaleData(seu_astrocytes, features = seu_astrocytes.genes)
seu_microglia <- NormalizeData(seu_microglia)
seu_microglia.genes <- rownames(seu_microglia)
seu_microglia <- ScaleData(seu_microglia, features = seu_microglia.genes)
seu_astrocytes@meta.data <- meta_astrocytes
seu_microglia@meta.data <- meta_microglia
Idents(seu_astrocytes = seu_astrocytes) <- seu_astrocytes@meta.data$'condition'
Idents(seu_microglia = seu_microglia) <- seu_microglia@meta.data$'condition'
```

```r
de_astrocytes <- FindMarkers(seu_astrocytes, ident.1 = "AD", ident.2 = "C")
de_microglia <- FindMarkers(seu_microglia, ident.1 = "AD", ident.2 = "C")
```
